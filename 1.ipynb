```python
import pandas_datareader as pdr
fred10 = pdr.get_data_fred('GS10')
```

```python
import pandas as pd
import numpy as np
```

```python
from bokeh.plotting import figure, show, output_notebook, curdoc
from bokeh.models import NumeralTickFormatter, Span

output_notebook()
curdoc().theme = "contrast"



plot = figure(height=400, 
              width=1000, 
              title="data haha",
              x_axis_label="date",
              x_axis_type="datetime",
              y_axis_label="gs10??",
              )

x = fred10.index
y = fred10["GS10"]

plot.line(x,y)
plot.yaxis.formatter = NumeralTickFormatter(format="0.00")

max = Span(
    location=fred10["GS10"].max(),
    dimension="width",
    line_color="red",
    line_width=1,
)
plot.add_layout(max)
min = Span(
    location=fred10["GS10"].min(),
    dimension="width",
    line_color="green",
    line_width=1,
)
plot.add_layout(min)

show(plot)
```
